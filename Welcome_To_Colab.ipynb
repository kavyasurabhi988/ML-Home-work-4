```
# ==============================
# PART C - Q1: NLP Preprocessing
# ==============================

!pip install nltk

import nltk

# ✅ FIXED downloads (ALL required)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')  # ✅ NEW FIX

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

text = "John enjoys playing football while Mary loves reading books in the library."

# Tokenization
tokens = word_tokenize(text)

# Stopword removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [w for w in tokens if w.lower() not in stop_words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(w) for w in filtered_tokens]

# POS tagging
pos_tags = pos_tag(lemmatized_tokens)

# Keep nouns & verbs
final_output = [word for word, tag in pos_tags if tag.startswith('N') or tag.startswith('V')]

print("Final Output:", final_output)
```

```
# ==========================================
# PART C - Q2: NER + Pronoun Ambiguity Check
# ==========================================

# Install spaCy
!pip install spacy

# Download English model
!python -m spacy download en_core_web_sm

# Import spaCy
import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "Chris met Alex at Apple headquarters in California. He told him about the new iPhone launch."

# Process text
doc = nlp(text)

# Step 1: Named Entity Recognition (NER)
print("Named Entities:\n")
for ent in doc.ents:
    print(f"{ent.text} --> {ent.label_}")

# Step 2: Pronoun Ambiguity Detection
pronouns = ["he", "she", "they", "him", "her", "them"]

found_pronoun = False
for token in doc:
    if token.text.lower() in pronouns:
        found_pronoun = True
        break

if found_pronoun:
    print("\n⚠️ Warning: Possible pronoun ambiguity detected!")
else:
    print("\nNo pronoun ambiguity detected.")
```
